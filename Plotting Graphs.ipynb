```python
nfl_attendance_df.plot(kind='bar', figsize=(8, 6), title='Home NFL Attendance by Team', y=['Home'])
```

```python
nfl_attendance_df.plot(kind='scatter', x='Home', y='Away', title='Home vs Away NFL Attendance', figsize=(8, 6))
print(nfl_attendance_df[['Team','Home', 'Away']])
```

```python
nfl_attendance_df['Total'].groupby(by=nfl_attendance_df['Division']).sum().plot(kind='pie', figsize=(8, 6), title='Average NFL Attendance by Division', autopct='%1.1f%%')
```

```python
nfl_attendance_df.set_index('Team').sort_values(by='Total', ascending=True)['Total'].plot(kind='barh', figsize=(10, 6), title='Total NFL Attendance by Team', xlabel='Attendance (in millions)', color='orange')
```

```python
plt.boxplot(nfl_attendance_df['Average'], vert=False)
plt.ylabel('Average Attendance')
plt.title('Boxplot of Average NFL Attendance')
```
